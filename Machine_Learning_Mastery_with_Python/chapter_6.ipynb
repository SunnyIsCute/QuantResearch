## Chapter 6 Understand Your Data With Visualization

#### 1. Univariate Plots

##### (1) Histograms

```python
# univariate histograms
import pandas as pd
import matplotlib.pyplot as plt
filename = 'data/pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pd.read_csv(filename, names=names)
data.hist(figsize=(10, 8))
plt.show()
```

We can see that perhaps the attributes age, pedi and test may have an exponential distribution. We can also see that perhaps the mass and pres and plas attributes may have a Gaussian or nearly Gaussian distribution. This is interesting because many machine learning techniques assume a Gaussian univariate distribution on the input variables.

##### (2) Density Plots

```python
# univariate density plots
data.plot(kind='density', figsize=(10,8), subplots=True, layout=(3,3), sharex=False)
plt.show()
```

##### (3) Box and Whisker Plots

Boxplots summarize the distribution of each attribute, drawing a line for
the median (middle value) and a box around the 25th and 75th percentiles (the middle 50% of the data). The whiskers give an idea of the spread of the data and dots outside of the whiskers show candidate outlier values (values that are 1.5 times greater than the size of spread of the middle 50% of the data).

```python
# Box and Whisker plots
data.plot(kind='box', figsize=(10,8), subplots=True, layout=(3,3), sharex=False, sharey=False)
plt.show()
```

#### 2. Multivariate Plots

##### (1) Correlation Matrix Plots

```python
# correlation matrix plot
import numpy as np

correlations = data.corr()

# plot correlation matrix
fig, ax = plt.subplots(figsize=(6, 6))
cax = ax.matshow(correlations, cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(len(correlations.columns))
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(correlations.columns, rotation=45, ha='left')
ax.set_yticklabels(correlations.columns)
plt.title("Correlation Matrix", pad=20)
plt.tight_layout()
plt.show()
```

##### (2) Scatter Plot Matrix

A scatter plot shows the relationship between two variables as dots in two dimensions, one axis for each attribute. You can create a scatter plot for each pair of attributes in your data.
Drawing all these scatter plots together is called a scatter plot matrix. Scatter plots are useful for spotting structured relationships between variables, like whether you could summarize the relationship between two variables with a line. Attributes with structured relationships may
also be correlated and good candidates for removal from your dataset.

```python
# scatterplot matrix
from pandas.plotting import scatter_matrix
scatter_matrix(data, figsize=(10,8))
plt.show()
```

Like the Correlation Matrix Plot above, the scatter plot matrix is symmetrical. This is useful to look at the pairwise relationships from dierent perspectives. Because there is little point of drawing a scatter plot of each variable with itself, the diagonal shows histograms of each attribute.
