#### 1.1 Daily Price

##### 1.1.1 Stock and Stock Index

```python
# !pip install yfinance
# !pip install mplfinance

import yfinance as yf
import mplfinance as mpf
from datetime import datetime
```

```python
start_date = datetime(2019, 1, 1)
end_date = datetime(2019, 12, 31)
data = yf.download('AAPL', start=start_date, end=end_date) # multi-index dataframe
```

```python
data.head()
```

```python
if isinstance(data.columns, pd.MultiIndex):
  data.columns = data.columns.droplevel(1) # drop the 'Ticker' level
```

```python
data.head()
```

```python
# Plot candlestick chart
mpf.plot(
    data,
    type='candle',
    mav=(3,6,9), # plot 3-day, 6-day and 9-day MAs to charts; each MA is computed using the 'Close' price
    volume=True,
    show_nontrading=True, # show non-trading days; useful for cryptocurrencies (traded 24/7) or to avoid visual gaps
    title='AAPL Stock Price',
    # style='yahoo',
    figratio=(10,6),
)
```

#### 2.1 Alphavantage

```python
import pandas as pd
import pandas_datareader as pdr

ts = pdr.av.time_series.AVTimeSeriesReader('AMZN', api_key='YOUR_FREE_API_KEY')
df = ts.read()
df.index = pd.to_datetime(df.index, format='%Y-%m-%d')
mpf.plot(df, type='candle', mav=(3,6,9), volume=True, show_nontrading=True)
```

#### 2.2 FRED

FRED has plenty of macro-economics data for example GDP, unemployment, inflation.

```python
import pandas_datareader as pdr
start = datetime(2019, 1, 1)
end = datetime(2019, 12, 31)
syms = ['DGS1MO', 'DGS3MO', 'DGS1', 'DGS3', 'DGS10']
df = pd.DataFrame()
for sym in syms:
  ts = pdr.fred.FredReader(sym, start=start, end=end)
  df1 = ts.read()
  df = pd.concat([df, df1], axis=1)

print(df)
```
